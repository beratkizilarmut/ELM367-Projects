##  Ayrık Zamanlı Fourier Dönüşümünü Hesaplanması, DTFT

```python
# Ayrık Zamanlı Fourier Dönüşümünü Hesaplanması, DTFT

w_cont = np.linspace(0,2*np.pi,1000)
X_abs = np.abs(2*np.cos(w_cont/2))
X_phase = np.array([-w/2 if w<np.pi else -w/2+np.pi for w in w_cont])
plt.subplot(121)
plt.plot(w_cont/np.pi, X_abs)
plt.title('$|X(e^{j\omega})|$')
plt.xlabel('Normalize frekans (x$\pi$)')
plt.subplot(122)
plt.plot(w_cont/np.pi, X_phase/np.pi)
plt.title('$ \measuredangle X(e^{j\omega})$')
plt.xlabel('Normalize frekans (x$\pi$)')
```

# Ayrık Fourier Dönüşümü Hesaplanması, DFT

```python
# Ayrık Fourier Dönüşümü Hesaplanması, DFT

def dirac(n):
    if n==0:
        return 1
    else:
        return 0
    
N = 8
X_k = np.zeros([N], dtype=complex)
for k in range(N):
    for n in range(N):
        X_k[k]  =X_k[k]+(dirac(n)+dirac(n-1))*np.exp(-1j*(2*np.pi/N)*k*n)
X_k_abs = np.abs(X_k)
X_k_phase = np.angle(X_k)
plt.figure()
plt.subplot(121)
plt.plot(w_cont/np.pi, X_abs, label='$|X(e^{j\omega})|$')
w_discrete = np.arange(0,N) * (2*np.pi/N)
plt.stem(w_discrete/np.pi, X_k_abs, 'r-', label='$|X[k]|$',markerfmt='ro')
plt.xlabel('$\omega$ (x$\pi$)')
plt.legend()
plt.subplot(122)
plt.plot(w_cont/np.pi, X_phase, label='$ \measuredangle X(e^{j\omega})$')
plt.stem(w_discrete/np.pi, X_k_phase, 'r-', label='$ \measuredangleX[k]$',markerfmt='ro')
plt.xlabel('$\omega$ (x$\pi$)')
plt.legend()
```

## DFT Genlik ve Faz Grafiklerinden İşaret Sentezi

```python
# DFT Genlik ve Faz Grafiklerinden İşaret Sentezi

x = np.array([0,2,0,-2])
N = len(x)
n = np.arange(0,4)
w_disc = n * 2 * np.pi / N
X_abs = np.abs(fft(x) / N)
X_phase = np.angle(fft(x) / N)
plt.figure()
plt.subplot(121)
plt.stem(w_disc / np.pi, X_abs)
plt.xlabel('$\omega (\\times \pi)$')
plt.title('$\\frac{1}{N}|X(e^{j\omega})|=|a_{k}|$')
plt.subplot(122)
plt.stem(w_disc / np.pi, X_phase / np.pi)
plt.xlabel('$\omega (\\times \pi)$')
plt.title('$\\frac{1}{N}\measuredangle X(e^{j\omega})(\\times\pi)=\measuredangle a_k(\\times \pi)$')
```

## a)

```python
# El Hesaplamaları
# İşarete parça parça bakacak olursak

# 𝛿[n] işaretinin fourier dönüşünümü 1 olur
dn = 1

# 𝛿[n-a] işaretinin, yani dirac delta function'ın fourier dönüşümü ise e üzeri (-jwa) olur 
# Elimizdeki işarette a = 1 olmaktadır, -1 değil.
w = np.linspace(0, 2*np.pi, 1000)
dn1 = np.exp(-1j * w * 1) 

# ℎ[𝑛] = 𝛿[𝑛] − 𝛿[𝑛 − 1]
hn = dn - dn1
```

```python
# Fourier dönüşümü alınmış işaretin fazı ve genliği hesaplandı ve çizdirildi

# Açısal Frekans Ayarlanması
w = np.linspace(0, 2*np.pi, 1000)

# H(e^jw) işaretinin fazı
h_phase = np.angle(hn)

# H(e^jw) işaretinin genliği
h_abs = np.abs(hn)

# Genlik ve Faz Grafiklerin çizilmesi
# Sürekli zamanlı işaret oldukları için plot kullanıldı
# Grafikle w'ya göre değil, w/pi'a göre çizdirilmiştir, dikkat ediniz.
plt.figure()
plt.subplot(121)
plt.plot(w/np.pi, h_abs)
plt.title('$|H[n]|-(Genlik) -FT$')
plt.xlabel('$\omega$ (x$\pi$)')
plt.subplot(122)
plt.plot(w/np.pi, h_phase)
plt.title('$ \measuredangle H[n] - (Faz) - FT$')
plt.xlabel('$\omega$ (x$\pi$)')
```

```python
n = np.arange(0, 11, 1)
w = np.pi / 3

# 1. Sorudaki dürtü yanıtı ile aynı olduğu için o adımda bulduğum fourier dönüşümünü direkt yazıyorum
h = 1 - np.exp(-1j * w * 1) 

# Not kısmında verildiği gibi çıkış işaretini bulmak için
# H(e^jw) işaretinin genliği ve fazı bulunmalı
h_phase = np.angle(h)
print('∠H(e^jw):',h_phase)
h_abs = np.abs(h)
print('|H(e^jw)|:',h_abs)

# Not kısmında verilen y formulunde değerler yerine konulursa
y = h_abs * np.cos(w * n + h_phase)

plt.figure()
plt.stem(n, y, use_line_collection=True)
plt.title('y[n] - Çıkış İşareti')
plt.xlabel('n')
```
