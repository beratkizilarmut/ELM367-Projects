# Örnek 4

```python
def signall(Fs,N,k): # Fs: Örnekleme frekansı
    # N : çizdirilmek istenen toplam örnek sayısı
    # k = x(t) işaretinin kaç periyot çizdirileceği
    F=100
    T=1/F
    t = np.linspace(-k*T/2,k*T/2,1000)
    xt = np.sinc(2*F*t)**2 # sürekli zamanlı işaret
    Ts=1/Fs
    n = np.arange(-N/2, N/2) # x[n] işaretinin indis ekseninin tanımlanması
    xn = np.sinc(2*F*n*Ts)**2 # x[n] işaretinin tanımlanması
    # x[n] işaretinin fourier transformu
    w = np.arange(-np.pi, np.pi, 2*np.pi/N) # omega ekseninin -pi ile +pi arasında tanımlanması
    xw = np.fft.fftshift(np.fft.fft(xn,N)/N) # ayrık zamanlı işaretin Fourier transformu
    plt.figure()
    plt.subplot(1,3,1)
    plt.plot(t,xt) # sürekli zamanlı x(t) işaretinin grafiğinin çizdirilmesi
    plt.xlabel("t (sn)") # grafiğin x ekseninin isimlendirilmesi
    plt.title("$x(t)$") # grafiğin y ekseninin isimlendirilmesi
    plt.subplot(1,3,2)
    plt.stem(n,xn) # x[n] işaretinin grafiğinin çizdirilmesi
    plt.xlabel("n (örnek)") # grafiğin x ekseninin isimlendirilmesi
    plt.title("$x[n]$") # grafiğin y ekseninin isimlendirilmesi
    plt.subplot(1,3,3)
    plt.stem(w/np.pi,abs(xw)) # X(w) işaretinin grafiğinin çizdirilmesi
    plt.xlabel("$\omega$ / $\pi$") # grafiğin x ekseninin isimlendirilmesi
    plt.title("$X(\omega)$ (Genlik)") # grafiğin y ekseninin isimlendirilmesi
    plt.show()
```

```python
signall(1000,30,3) # Fs= 1000 Hz , N = 30 örnek , x(t) toplam 3 periyot için çizdirildi
```

```python
signall(600,30,5) # Fs= 600 Hz , N = 30 örnek , x(t) toplam 5 periyot içi çizdirildi
```

```python
signall(300,30,10) # Fs= 300 Hz , N = 30 örnek , x(t) toplam 10 periyotta çizdirildi
```

```python
signall(100,30,30) # Fs= 100 Hz , N = 30 örnek , x(t) toplam 3 periyotta çizdirildi
```

```python
F=100 # x işaretinin frekansı
T=1/F # x işaretinin periyodu
k=2 # Periyot sayısı
t=np.arange(0.,k*T,0.00001) # x işaretine ait t değişkeninin tanımlanması
x=5*np.cos(200*np.pi*t) # x(t) işaretinin tanımlanması
plt.figure() # yeni bir figure penceresi açar
plt.xlabel("t") # grafiğin x ekseninin isimlendirilmesi
plt.title("x(t)") # Grafiğin başlığının oluşturulması
plt.plot(t,x) # işaretinin grafiğinin çizdirilmesi
plt.show() # grafiğin gösterilmesi
```

```python
Fs=2200 # örnekleme frekansının tanımlanması
Ts=1/Fs # örnekleme periyodunun tanımlanması
N=22 # bir periyottaki örnek sayısının tanımlanması
n=np.arange(0.,k*N) # örnekleme indisinin 0'dan k periyot olacak şekilde array olarak tanımlanması
xn=5*np.cos(200*np.pi*n*Ts) #örneklenmiş x[n] işaretinin tanımlanması
```

```python
def quantize(x, S):
    X = x.reshape((-1,1))
    S = S.reshape((1,-1))
    dists = abs(X-S)
    nearestIndex = dists.argmin(axis=1)
    quantized = S.flat[nearestIndex]
    return quantized.reshape(x.shape)
```

```python
k_bit= 2 # işaret değerlerinin kaç bit ile ifade edileceği
Max_range= np.amax(xn)-np.amin(xn) # işaretin y ekseninde aldığı max değer aralığı
q=pow(2,k_bit) # seviye sayısı
delta = Max_range/q # adım aralığı
S_midrise = -np.amax(xn) + delta/2 + np.arange(q)*delta # mid-rise yöntemi için formül
S_midtread = -np.amax(xn) + np.arange(q)*delta # mid-tread yöntemi için formül
```

```python
y_midtrise = quantize(xn, S_midrise) # mid-rise yöntemi ile kuantalanmış değerler
y_midtread = quantize(xn, S_midtread) # mid-tread yöntemi ile kuantalanmış değerler
```

```python
plt.figure() #yeni bir figure penceresi açar
fig, ax = plt.subplots()
ax.stem(n,xn) # x[n] işaretinin grafiğinin çizdirilmesi
ax.bar(n,y_midtrise,color='orange',edgecolor = "darkorange") # kuantalanmış işaretin grafiğinin çizdirilmes
plt.xlabel("örnek") # grafiğin x ekseninin isimlendirilmesi
plt.title('midrise yöntemiyle kuantalanmış işaret')
fig, ax = plt.subplots()
ax.stem(n,xn) # x[n] işaretinin grafiğinin çizdirilmesi
ax.bar(n,y_midtread,color='orange',edgecolor = "darkorange") # kuantalanmış işaretin grafiğinin çizdirilmes
plt.xlabel("örnek") # grafiğin x ekseninin isimlendirilmesi
plt.title('midtread yöntemiyle kuantalanmış işaret')
plt.show() # grafiklerin gösterilmesi
```

## b) İşaretlerin Frekans Domaininde İncelenmesi

```python
Fs = 2200 # Örnekleme Frekansı Tanımlandı
Ts = 1/Fs # Örnekleme Periyodu Tanımlandı
N = 22 # Örnek Sayısı Tanımlandı
n = np.arange(0,k*N) # İki Periyot Array Tanımlandı
x1 = 7 * np.cos(200*np.pi*n*Ts) # x1 İşareti Tanımlandı
x2 = 3 * np.cos(200*np.pi*n*Ts) + 4 * np.cos(1800*np.pi*n*Ts) # x2 İşareti Tanımlandı

# Fourier Dönüşümleri Alındı
w = np.arange(-np.pi, np.pi, 2*np.pi/N)
x1f = np.fft.fftshift(np.fft.fft(x1,N)/N)
x2f = np.fft.fftshift(np.fft.fft(x2,N)/N)

# Genlik Grafiklerin Çizdirilmesi
plt.figure()
plt.subplot(2,1,1)
plt.title('$X1(\omega)$ ve $X2(\omega)$ İşaretleri Genlik Grafikleri')
plt.stem(w/np.pi, np.abs(x1f))
plt.xlabel("$\omega$ / $\pi$")
plt.ylabel("$X1(\omega)$ (Genlik)")
plt.subplot(2,1,2)
plt.stem(w/np.pi, np.abs(x2f))
plt.xlabel("$\omega$ / $\pi$")
plt.ylabel("$X2(\omega)$ (Genlik)")
plt.show()

# Faz Grafiklerin Çizdirilmesi
plt.figure()
plt.subplot(2,1,1)
plt.title('$X1(\omega)$ ve $X2(\omega)$ İşaretleri Faz Grafikleri')
plt.stem(w/np.pi, np.angle(x1f))
plt.xlabel("$\omega$ / $\pi$")
plt.ylabel("$X1(\omega)$ (Faz)")
plt.subplot(2,1,2)
plt.stem(w/np.pi, np.angle(x2f))
plt.xlabel("$\omega$ / $\pi$")
plt.ylabel("$X2(\omega)$ (Faz)")
plt.show()
```

### Genlik grafiğine baktığımızda iki işaret içinde pi/11 noktasında dirac'lar bulunmaktadır fakat değerleri beklenildiği gibi katsayı farkından dolayı farklıdır. ## c) Genlik ve Faz Grafiklerinden Reconstruction İşlemi

### x1 işareti için, genlik grafiğine bakıldığında π/11 ve -π/11 noktalarında elementleri olduğu gözlenir. Bu elementler işaretin ω'sunu verir. Bize gerekli olan Ω değerini almak için Ω = ω * fs eşitliğinden Ω = 200 π rad/sn olarak bulunur.  Faz grafiğine bakıldığında bileşen bulunan noktalara bakıldığında faz grafiğinde sıfır değeri aldığı için işaretin cosinus olduğu öğrenilir. Genlik grafiğinde tepe noktası 3.5 olarak bulunur, işaretin cosinus olduğunu biliyoruz, cosinus'ün ters fourier dönümüşünde 2 ile çarpmamız gerektiği için genlik 7 olarak bulunur. Bu bilgiler ışığında işaret x1 = 7 cos ( 200πt ) olur.

```python
Fs = 2200 # Örnekleme Frekansı Tanımlandı
Ts = 1/Fs # Örnekleme Periyodu Tanımlandı
N = 22 # Örnek Sayısı Tanımlandı
n = np.arange(0,k*N) # İki Periyot Array Tanımlandı
x1r = 7 * np.cos(200*np.pi*n*Ts) # x1 İşareti Tanımlandı

# Grafik Çizdirilmesi
plt.figure()
plt.stem(n,x1r)
plt.title("x1r[n], Fs=2200 Hz")
plt.xlabel("Örnek")
plt.ylabel("x1[n]")
plt.show()
```

### x2 işareti için, genlik grafiğine bakıldığında π/11 ve -π11 noktalarında bir çift,  9π/11 ve -9π/11 noktalarında da ikinci bir çift element olduğu gözlenir. Bu elementler işaretin ω'sunu verir. ## b) İşaretlerin Frekans Domaininde İncelenmesi

```python
Fs = 1000 # Örnekleme Frekansı Tanımlandı
Ts = 1/Fs # Örnekleme Periyodu Tanımlandı
N = 10 # Örnek Sayısı Tanımlandı
n = np.arange(0,k*N) # İki Periyot Array Tanımlandı
x1 = 7 * np.cos(200*np.pi*n*Ts) # x1 İşareti Tanımlandı
x2 = 3 * np.cos(200*np.pi*n*Ts) + 4 * np.cos(1800*np.pi*n*Ts) # x2 İşareti Tanımlandı

# Fourier Dönüşümleri Alındı
w = np.arange(-np.pi, np.pi, 2*np.pi/N)
x1f = np.fft.fftshift(np.fft.fft(x1,N)/N)
x2f = np.fft.fftshift(np.fft.fft(x2,N)/N)

# Genlik Grafiklerin Çizdirilmesi
plt.figure()
plt.subplot(2,1,1)
plt.title('$X1(\omega)$ ve $X2(\omega)$ İşaretleri Genlik Grafikleri')
plt.stem(w/np.pi, np.abs(x1f))
plt.xlabel("$\omega$ / $\pi$")
plt.ylabel("$X1(\omega)$ (Genlik)")
plt.subplot(2,1,2)
plt.stem(w/np.pi, np.abs(x2f))
plt.xlabel("$\omega$ / $\pi$")
plt.ylabel("$X2(\omega)$ (Genlik)")
plt.show()

# Faz Grafiklerin Çizdirilmesi
plt.figure()
plt.subplot(2,1,1)
plt.title('$X1(\omega)$ ve $X2(\omega)$ İşaretleri Faz Grafikleri')
plt.stem(w/np.pi, np.angle(x1f))
plt.xlabel("$\omega$ / $\pi$")
plt.ylabel("$X1(\omega)$ (Faz)")
plt.subplot(2,1,2)
plt.stem(w/np.pi, np.angle(x2f))
plt.xlabel("$\omega$ / $\pi$")
plt.ylabel("$X2(\omega)$ (Faz)")
plt.show()
```

### Bir önceki kısımda da söylediğimiz gibi, aliasingden dolayı iki işaretin genlik ve faz (grafikler birebir aynı olmasa bile bizi ilgilendiren kısımlar yani element bulunan kısımlarda) grafikleri aynı oluşmuştur.

```python
xn = np.array([1.25, 3.5, 4, 6.7, -1.4, -3.5, -10.5, 8.5, 3.8])
n=np.arange(0, 9 ,1)

def quantize(x, S):
    X = x.reshape((-1,1))
    S = S.reshape((1,-1))
    dists = abs(X-S)
    nearestIndex = dists.argmin(axis=1)
    quantized = S.flat[nearestIndex]
    return quantized.reshape(x.shape)

k_bit = 3 # Bit Miktarı
Max_range = np.amax(xn)-np.amin(xn) # Değer aralığı genişliği
q = pow(2,k_bit) # Seviye Sayısı
delta = Max_range/q # Adım Aralığı

S_midrise = -np.amax(xn) + delta/2 + np.arange(q)*delta # Mid-Rise Formulü
S_midtread = -np.amax(xn) + np.arange(q)*delta # Mid-Tread Formulü

y_midtrise = quantize(xn, S_midrise) # Mid-Rise Yöntemi ile kuantizasyon
y_midtread = quantize(xn, S_midtread) # Mid-Tread Yöntemi ile kuantizasyon


# Grafiklerin Çizdirilmesi
fig, ax = plt.subplots()
ax.stem(n,xn) 
ax.bar(n,y_midtrise,color='orange',edgecolor = "darkorange") 
plt.xlabel("Örnek") 
plt.title('x[n], k = 3 bit, Mid-Rise Quantized')

fig, ax = plt.subplots()
ax.stem(n,xn) 
ax.bar(n,y_midtread,color='orange',edgecolor = "darkorange")
plt.xlabel("Örnek")
plt.title('x[n], k = 3 bit, Mid-Tread Quantized')
plt.show()
```
