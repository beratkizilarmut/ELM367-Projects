```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

# Soru 7)

# Soru 2.34. de ki sistemin genlik ve fazı çizdirildi

```python
# Frekans aralığı ve işaret sistemi tanımlandı
w = np.linspace(-np.pi, np.pi, 50)
H = 1 - ( 0.45 * np.exp(-1j*w) / (1 - 0.8*np.exp(-1j*w) ) )

# Normal durumların aksine, stem yerine plot olarak grafikler çizdirildi
# Bunun sebebi sistemin ayrık zamanlı olmaması
plt.figure()
plt.title("H(e^jw) System, Magnitude")
plt.plot(w, np.abs(H))
plt.show()
```

```python
plt.figure()
plt.title("H(e^jw) System, Phase")
plt.plot(w, np.angle(H))
plt.show()
```

# Soru 8)

# Soru 2.32. de ki sistemin genlik ve fazı a = -0.5 için çizdirildi

```python
# İşaret sistemi, aralık ve a tanımlandı
a = -0.5
w = np.linspace(-np.pi, np.pi, 50)
X = 1 / (1 - a * np.exp(-1j*w) )

# Normal durumların aksine, stem yerine plot olarak grafikler çizdirildi
# Bunun sebebi sistemin ayrık zamanlı olmaması
plt.figure()
plt.title("X(e^jw) System, Magnitude")
plt.plot(w, np.abs(X))
plt.show()
```

```python
plt.figure()
plt.title("X(e^jw) System, Phase")
plt.plot(w, np.angle(X))
plt.show()
```

# Soru 9)

# Soru 2.36. de ki sistemin genlik ve fazı çizdirildi
