```python
# Hazır komut kullanmadan konvolusyon hesaplanması

h=np.array([1,2,-1],dtype=float)
x=np.array([1,3],dtype=float)
M=len(h)
N=len(x)
L=M+N-1
y=np.zeros([L])
for n in range(L):
    y[n]=0
    for k in range(N):
        if n-k<0 or n-k>=M:
            h_n_k=0
        else:
            h_n_k=h[n-k]
        y[n]=y[n]+x[k]*h_n_k
print(y)
```

```python
# Hazır komut kullanarak konvolusyon hesaplanması

from scipy import signal
h=np.array([1,2,-1],dtype=float)
x=np.array([1,3],dtype=float)
print(np.convolve(x,h))
print(signal.convolve(x,h))
```

```python
# Birim dürtü yanıtları -5 den 30'a olacak şekilde tanımlandı
h1 = np.zeros(40)
h2 = np.zeros(40)
h1[0:9] = [0, 0, 0, 0, 0.33, 0.33, 0.33, 0, 0]
h2[0:9] = [0, 0, 0, 0, 0, 1, 0, 0, -1]

# -5'den 30'a kadar aralık tanımlandı
n = np.arange(-5, 30, 1)

# Unit Step Functionlar variable olarak hazırlandı
# Function tanımları içerisinde direkt yazıldığında sorun çıkmaktaydı
u = np.heaviside(n, 1)
u5 = np.heaviside(n-5, 1)
u20 = np.heaviside(n-20, 1)
```

```python
# Giriş 1 tanımlandı
x1 = np.cos(np.pi * n / 5) * (u - u20)
plt.figure()
plt.stem(n,x1, use_line_collection=True)
plt.title('x1[n] input itself')
plt.show()
```

```python
# Giriş 2 tanımlandı
x2 = (-0.3)**(n) * (u - u5)
plt.figure()
plt.stem(n,x2, use_line_collection=True)
plt.title('x2[n] input itself')
plt.show()
```
