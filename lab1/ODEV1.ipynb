```python
# Ötelenecek x[n] ayrık zamanlı işareti
n=np.array([0,1,2,3,4])
x3=np.array([1,0,2,-1,0])
plt.figure()
plt.xlabel('n')
plt.ylabel('x[n]')
plt.stem(n,x3)
plt.show()
```

```python
# Ötelenmiş olan, x[n-2] ayrık zamanlı işareti
n=np.array([0,1,2,3,4])
# N dizisindeki bütün elemanlara 2 eklenir
n2=n+2
x3=np.array([1,0,2,-1,0])
plt.figure()
plt.title('x[n-2]')
plt.stem(n2,x3)
plt.show()
```

```python
# Öteleme işlemi ikinci yol
n=np.array([0,1,2,3,4,5])
x=np.array([1,0,2,-1,0,0])
# Öteleme fonksiyonu ile öteleme işlemi
x_shifted=np.append(np.array([0,0]),x[:-2])
fig = plt.figure()
ax1=fig.add_subplot(211)
ax2=fig.add_subplot(212)
ax1.stem(n,x)
ax1.title.set_text('x[n]')
ax2.stem(n,x_shifted)
ax2.title.set_text('x[n-2]')
```

# SORU-1

```python
# Kütüphane importlanması
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

# Istenilen değişkenler tanımlanıldı
a = 3 - 4j
b = 1 + 2j
C = a + b
D = a / b
```

## I)

```python
# C ve D işaretlerinin .real işlemi ile gerçek kısımları çağırılıp, yazdırıldı
print('C Signal, Real Part = %.1f' % (C.real))
print('D Signal, Real Part = %.1f' % (D.real))
```

## II)

```python
# C ve D işaretlerinin .imag işlemi ile imajiner kısımları çağırılıp, yazdırıldı
print('C Signal, Imaginary Part = %.1f' % (C.imag))
print('D Signal, Imaginary Part = %.1f' % (D.imag))
```

# III)

```python
# C ve D işaretlerinin abs işlemi ile genliği yazdırıldı
print('C Signal, Amplitude = %.3f' % (np.abs(C)))
print('D Signal, Amplitude = %.3f' % (np.abs(D)))
```

# IV)

```python
# C ve D işaretlerinin angle işlemi ile fazı yazdırıldı
print('C Signal, Phase = %.3f' % (np.angle(C)))
print('D Signal, Phase = %.3f' % (np.angle(D)))
```

# SORU-3

```python
# Kütüphane importlanması
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

# I)

```python
# İşaret n'e bağlı olmadığı için işaret sabit bir değer aldı
# e üzeri (pi/5) değerinde sabit kaldı
```

```python
# n için, birer artış ile 0-15 aralığı
n = np.arange(0,16,1)
# n ile aynı uzunlukta, birler dizisi oluşturuldu
x1 = np.ones(16)
# bu dizi verilen sayı ile çarpıldı
x1 = x1 * np.exp(np.pi / 5)
plt.figure()
# Ayrık zamanlı x1 işaretinin yazdırılması
plt.stem(n,x1)
plt.title('x1[n] Signal')
plt.show()
```

# II)

```python
# İlk kısımda da olduğu gibi işaret n'e bağlı olmadığı için n'e göre plotlandığında sabit oldu
# fakat bu sefer argümanın içinde j'de olduğu için sonuç karmaşık sayı oldu
```

```python
# n için, birer artış ile 0-15 aralığı
n = np.arange(0,16,1)
# n ile aynı uzunlukta, birler dizisi oluşturuldu
x2 = np.ones(16, dtype=complex)
# bu dizi verilen sayı ile çarpıldı
x2 = x2 * np.exp(np.pi * 1j / 5)
# Ayrık zamanlı x2 işaretinin yazdırılması
# Kompleks olduğu için iki ayrı grafik yazdırılacak
fig = plt.figure()
# Subplotlar arası mesafe eklenildi
plt.subplots_adjust(hspace=1)
ax1=fig.add_subplot(211)
ax2=fig.add_subplot(212)
# Subplotlara yazıldı
ax1.stem(n,x2.real)
ax1.title.set_text('x2[n] Signal, Real Part')
ax2.stem(n,x2.imag)
ax2.title.set_text('x2[n] Signal, Imaginary Part')
```

# III)

```python
# İşaret, euler açılımından da beklediğimiz gibi sinüzodial oldu
# Real ve imag kısımları arasında faz farkı oluştu
```

```python
# n için, birer artış ile 0-15 aralığı
n = np.arange(0,16,1)
x3 = np.exp(np.pi * 1j * n / 5)
# Ayrık zamanlı x3 işaretinin yazdırılması
# Kompleks olduğu için iki ayrı grafik yazdırılacak
fig = plt.figure()
# Subplotlar arası mesafe eklenildi
plt.subplots_adjust(hspace=1)
ax1=fig.add_subplot(211)
ax2=fig.add_subplot(212)
# Subplotlara yazıldı
ax1.stem(n,x3.real)
ax1.title.set_text('x3[n] Signal, Real Part')
ax2.stem(n,x3.imag)
ax2.title.set_text('x3[n] Signal, Imaginary Part')
```

# IV)

```python
# Euler dönüşümüne dikkate alınırsa, x4 işareti ile x2 işareti aynı işaret olmaktadır
# Bu işaret de sabit bir değerde kalacaktır
```

```python
# n için, birer artış ile 0-15 aralığı
n = np.arange(0,16,1)
# n ile aynı uzunlukta, birler dizisi oluşturuldu
x4 = np.ones(16, dtype=complex)
x4 = x4 * np.cos(np.pi/5) + (1j*np.sin(np.pi/5))
# Ayrık zamanlı x4 işaretinin yazdırılması
# Kompleks olduğu için iki ayrı grafik yazdırılacak
fig = plt.figure()
# Subplotlar arası mesafe eklenildi
plt.subplots_adjust(hspace=1)
ax1=fig.add_subplot(211)
ax2=fig.add_subplot(212)
# Subplotlara yazıldı
ax1.stem(n,x4.real)
ax1.title.set_text('x4[n] Signal, Real Part')
ax2.stem(n,x4.imag)
ax2.title.set_text('x4[n] Signal, Imaginary Part')
```

# V)

```python
# Bir önceki adımda olduğu gibi euler dönüşümüne dayanarak, x3 ile x5 işaretlerinin aynı olduğu söylenebilir
```

```python
# n için, birer artış ile 0-15 aralığı
n = np.arange(0,16,1)
x5 = np.cos(np.pi * n/5) + (1j*np.sin(np.pi * n/5))
# Ayrık zamanlı x5 işaretinin yazdırılması
# Kompleks olduğu için iki ayrı grafik yazdırılacak
fig = plt.figure()
# Subplotlar arası mesafe eklenildi
plt.subplots_adjust(hspace=1)
ax1=fig.add_subplot(211)
ax2=fig.add_subplot(212)
# Subplotlara yazıldı
ax1.stem(n,x5.real)
ax1.title.set_text('x5[n] Signal, Real Part')
ax2.stem(n,x5.imag)
ax2.title.set_text('x5[n] Signal, Imaginary Part')
```
